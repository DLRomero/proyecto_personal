```python
qualys= pd.read_csv('./Data/qualifying.csv')
qualys.head(2)
```

```python
def limpieza_merge(data_original, drivers, constructors, races):
    data_original= data_original[data_original['raceId']> 1073]
    df1= pd.merge(data_original, drivers, how='left', on='driverId')
    df2= pd.merge(df1, constructors, how='left', on='constructorId')
    df3= pd.merge(df2, races, how='left', on='raceId')
    df3.rename(columns={'name_x': 'driver_name', 'name_y': 'constructor'}, inplace=True)
    nuevas_columnas = {col: col.lower() for col in df3.columns}
    df3.rename(columns = nuevas_columnas, inplace = True)

    return df3
```

```python
qualys_info= limpieza_merge(qualys, drivers, constructors, races)
qualys_info.head()
```

```python
qualys_info.drop(columns=['number_x', 'number_y'], inplace=True)

qualys_info.head()
```

```python
qualys_info.head()
```

```python
qualys_info.rename(columns={'name': 'circuit'}, inplace=True)
```

```python
qualys_info.to_csv('./Data/qualys_resume.csv')
```

```python
results= pd.read_csv('./Data/results.csv')
results.head()
```

```python
results_info= limpieza_merge(results, drivers, constructors, races)
```

```python
results_info.drop(columns=['number_y', 'number_x', 'time_y'], inplace=True)
results_info.head(30)
```

```python
results_info['overtakes']= results_info['grid'] - results_info['positionorder']
```

```python
results_info.rename(columns={'time_x': 'time', 'name': 'circuit'}, inplace=True)
results_info.head()
```

```python
results_info.to_csv('./Data/results_resume.csv')
```
